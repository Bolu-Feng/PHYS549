```python
from numba import jit 

import numpy as np
import matplotlib.pyplot as plt
```

```python
 # Plot attribute settings
 plt.rc("xtick", labelsize="large")
 plt.rc("ytick", labelsize="large")
 plt.rc("axes", labelsize="xx-large")
 plt.rc("axes", titlesize="xx-large")
 plt.rc("figure", figsize=(8,8))
```

```python
# constants 
mu_0 = np.pi * 4.0 * pow(10, -7) # permeability of free space [kg*m*s^-2*A^-2]
eps_0 = 8.854187817e-12
q_p = 1.602176634e-19         # proton charge [coulombs]
m_p = 1.6726219236951e-27          # proton mass [kg]
m_e = 9.1093837e-31
#q_p = 1.6022e-19         # proton charge [coulombs]
#m_p = 1.6726e-27          # proton mass [kg]
mu = 10000000.0 * np.array([0.0, 0.0, 1.0]) # set magnetic moment to point in z direction
N = 100
```

```python
def bot_field(x,y,z,t,dt,beta):
    z_a = 1000 - beta*3e8*t 
    z_b = 1000
    beta_v = np.array([0.0, 0.0, -beta]) 
    # point dipole A
    pos_A = np.array([0.0, 0.0, z_a])        # set the position of the first dipole
    r_A = np.array([x,y,z]) - pos_A             # find the difference between this point and point of the observer
    rmag_A = np.sqrt(sum(r_A**2))
    factor = 1 - np.dot((r_A/rmag_A),beta_v)
    B1_A = 3.0*r_A*np.dot(mu,r_A) / (rmag_A**5) /factor  # calculate the first term to the magnetic field
    B2_A = -1.0 * mu / (rmag_A**3) /factor

    #E
    z_a_1 = 3e5 - beta*3e8*(t-dt) 
    pos_A_1 = np.array([0.0, 0.0, z_a_1])        # set the position of the first dipole
    r_A_1 = np.array([x,y,z]) - pos_A_1             # find the difference between this point and point of the observer
    rmag_A_1 = np.sqrt(sum(r_A_1**2))
    factor_1 = 1 - np.dot((r_A_1/rmag_A_1),beta_v)
    
    E_1 = r_A_1/(rmag_A_1**3*factor_1)

    z_a_2 = 3e5 - beta*3e8*(t+dt) 
    pos_A_2 = np.array([0.0, 0.0, z_a_2])        # set the position of the first dipole
    r_A_2 = np.array([x,y,z]) - pos_A_2             # find the difference between this point and point of the observer
    rmag_A_2 = np.sqrt(sum(r_A_2**2))
    factor_2 = 1 - np.dot((r_A_2/rmag_A_2),beta_v)
    E_2 = r_A_2/(rmag_A_2**3*factor_2)

    E = 1/(4.0*np.pi*eps_0)*(E_2 - E_1)/(2*dt)



    
    # point dipole B
    pos_B = np.array([0.0, 0.0, -z_b])  # set the position of the first dipole
    r_B = np.array([x,y,z]) - pos_B        # find the difference between this position and the observation position
    rmag_B = np.sqrt(sum(r_B**2))
    B1_B = 3.0*r_B*np.dot(mu,r_B) / (rmag_B**5) # calculate the first term to the magnetic field
    B2_B = -1.0 * mu / (rmag_B**3)              # calculate the second term

    B = ((mu_0/(4.0*np.pi)) * (B1_A + B2_A + B1_B + B2_B))

    
    return B, E
```

```python
B, E = bot_field(1,1,1, 1, 1e-3, 0.01) 
```

```python
y = np.arange(-3e5, 3e5, 1e4) # create a grid of points from y = -10 to 10
z = np.arange(-3e5, 3e5, 1e4) # create a grid of points from z = -10 to 10
Y, Z = np.meshgrid(y,z)        # create a rectangular grid out of y and z
len_i, len_j = np.shape(Y)       # define dimensions, for use in iteration
Bf = np.zeros((len_i,len_j,3))
Ef = np.zeros((len_i,len_j,3))   # initialize all points with zero magnetic field
```

```python
for i in range(0, len_i): 
    for j in range(0, len_j):
        Bf[i,j], Ef[i,j] = bot_field(0.0, Y[i,j], Z[i,j], 0, 0, 0) 
        
plt.streamplot(Y,Z, Bf[:,:,1], Bf[:,:,2], color='orange') # plot the magnetic field
plt.xlim(-3e5,3e5)
plt.ylim(-3e5,3e5)
plt.xlabel("$y$ (m)")
plt.ylabel("$z$ (m)")
plt.title("Magnetic Field in a Magnetic Bottle")
```

```python
q = q_p # charge of helium-4
m = m_p # mass of helium-4
QoverM = q/m
beta = 0.01

dt = pow(10, -4) # timestep

t = np.arange(0.0, 1, dt) # array for times
rp = np.zeros((len(t), 3)) # array for position values
vp = np.zeros((len(t), 3)) # array for velocity values

v_o = 1e5 # set the initial velocity to 100 m/s
rp[0,:] = np.array([1e5, 1e5, 0.0]) # initialize the position to y=-5, 5m above the lower dipole
vp[0,:] = np.array([0.0, 0.0, v_o]) # initialize the velocity to be in the z-direction
```

```python
for it in np.arange(0, len(t)-1,1):
    t_now = it*dt
    v_now = (sum(vp[it]**2))**0.5
    if it%1000 == 0:
      print(t_now, v_now)
    
    B, E = bot_field(rp[it,0], rp[it, 1], rp[it,2], t_now, dt, beta)
    j1 = (QoverM * (np.cross(vp[it,:], B) + E) )*dt        
    k1 = dt*vp[it]
    vp_k1 = vp[it] + 0.5*j1
    rp_k1 = rp[it] + 0.5*k1
  
    B, E = bot_field(rp_k1[0], rp_k1[1], rp_k1[2], t_now+0.5*dt, dt, beta)
    j2 = (QoverM * (np.cross(vp_k1, B) + E))*dt        
    k2 = dt*(vp[it] + 0.5*j1)
    vp_k2 = vp[it] + 0.5*j2
    rp_k2 = rp[it] + 0.5*k2

    B, E = bot_field(rp_k2[0], rp_k2[1], rp_k2[2], t_now+0.5*dt, dt, beta)
    j3 = (QoverM * (np.cross(vp_k2, B) + E))*dt        
    k3 = dt*(vp[it] + 0.5*j2)
    vp_k3 = vp[it] + j3
    rp_k3 = rp[it] + k3

    B, E = bot_field(rp_k3[0], rp_k3[1], rp_k3[2], t_now+dt, dt, beta)
    j4 = (QoverM * (np.cross(vp_k3, B) + E))*dt        
    k4 = dt*(vp[it] + j3)

    vp[it+1] = vp[it] + 1.0/6.0*(j1 + 2*j2 + 2*j3 + j4)                     # update the velocity of the particle based on this force
    rp[it+1] = rp[it] + 1.0/6.0*(k1 + 2*k2 + 2*k3 + k4)                # update the positron of the particle based on this velocity
    #if (np.sqrt(np.sum(rp[it+1]**2)) > 500.0): # If the particle escapes (i.e. exceeds 20 m from origin), cease calculations
     #   break
```
